# **1. Effect-Free Programming**

<p align="justify"> <b>Effect-free programming</b> adalah pendekatan dalam pemrograman yang menekankan <i>pada penulisan kode yang tidak menyebabkan efek samping</i>. Dalam konteks pemrograman fungsional, effect-free programming sangat penting karena mendukung prinsip-prinsip utama dari paradigma ini.

<p align="justify"> Dalam konsep matematika, fungsi adalah relasi antara dua himpunan yang memetakan setiap elemen dari himpunan pertama (domain) ke elemen himpunan kedua (range). Fungsi biasanya dinotasikan sebagai f(x), di mana x adalah input dan f(x) adalah output.

Contoh sederhana: f(x)=x^2

<p align="justify"> Dalam fungsi ini, jika kita memberi input x=3, maka outputnya akan selalu 9. Hal ini disebut <b>deterministik</b>, yaitu <i>fungsi selalu menghasilkan output yang sama untuk input yang sama</i>. Selain itu, fungsi matematika f(x)=x2 <i>tidak mengubah nilai atau variabel lain di luar perhitungannya</i>. Hasil perhitungannya hanya akan bergantung pada input x dan tidak ada perubahan state yang terjadi. Hal ini disebut juga <b>Effect-free Programming,</b> di mana fungsi tidak mengubah variabel global, tidak melakukan operasi input/output, dan tidak mengubah data di luar ruang lingkupnya.


# **2. Pure Function**

<p align="justify"> Untuk menciptakan sebuah kode yang menerapkan pendekatan Effect-free programming, kita dapat menggunakan teknik Pure Function. <b>Pure function</b> adalah sebuah fungsi yang selalu mengembalikan/me-return output yang sama ketika diberi input yang sama (sebagaimana contoh fungsi matematika
sebelumnya). Artinya, untuk setiap inputan, fungsi ini akan <b>mengembalikan hasil yang sama secara konsisten</b>, tidak peduli kapan atau dimana fungsi dipanggil.
Selain itu, pure function <b>tidak menyebabkan efek samping apa pun</b> selain
me-return value. Artinya, pure function tidak memodifikasi state dalam sistem di
luar scope-nya. Mirip-mirip dengan konsep Effect-free programming bukan? Yup, hal ini dikarenakan Pure Function merupakan implementasi dari Effect-free
programming. Perhatikan contoh berikut:



```python
def tambah(a, b):
  return a + b

hasil =  tambah(3, 4)
print("Jumlah: ", hasil)
```

<p align="justify"> Dalam contoh di atas, fungsi tambah merupakan pure function. Fungsi tambah selalu menghasilkan output yang sama untuk input yang sama. Jika kita memanggil 'tambah( 3, 4 )',  hasilnya akan selalu '7', tidak peduli berapa kali atau kapan fungsi ini dipanggil.

<p align="justify"> Selain itu, fungsi 'tambah' tidak memodifikasi state dalam sistem diluar scope. Fungsi ini hanya menerima dua argumen, yaitu a dan b, menjumlahkannya, dan mengembalikan hasilnya. <b> Fungsi ini tidak mengubah variabel global, tidak melakukan input-output </b>, atau memodifikasi data di luar ruang lingkupnya.

<p align="justify"> Dalam contoh lain, kita memiliki fungsi 'total_nilai' yang akan menghitung total nilai dari data yang sudah diberikan. Fungsi ini hanya bergantung pada parameter yang dibutuhkan, yaitu 'nilai'. Perhatikan contoh berikut:

```python
nilai = [
    {'matkul': 'Fungsional', 'nilai': 95},
    {'matkul': 'Mobile', 'nilai': 55}
]

def total_nilai (nilai):
  total_awal = 0
  for item in nilai:
    total_awal += item['nilai']

  return total_awal

hasil_nilai = total_nilai(nilai)
print("Total nilai: ", hasil_nilai)
```

<p align="justify"> Selanjutnya, kita memiliki fungsi 'tambah_nilai' yang digunakan untuk
menambahkan nilai matkul lain ke dalam kumpulan nilai yang telah kita punya.
Fungsi ini akan memodifikasi variabel nilai yang ada di luar fungsi.

```python
def tambah_nilai(nama_matkul, jumlah_nilai):
    global nilai
    for item in nilai:
        if item['matkul'] == nama_matkul:
            item['nilai'] += jumlah_nilai
            print(f"Nilai {nama_matkul} ditambahkan {jumlah_nilai}. \nTotal nilai {nama_matkul} saat ini {item['nilai']}")
            break
    else:
        print(f"Mata kuliah {nama_matkul} tidak ditemukan dalam daftar!")

# Cek nilai awal
print("Nilai awal: ",nilai)

# Menambahkan nilai ke dalam daftar
tambah_nilai('Mobile', 15)

# Cek nilai saat ini
print("Nilai update:",nilai)
```

<p align="justify"> Fungsi 'tambah_nilai' <b>bukan</b> merupakan fungsi yang murni (pure function) karena fungsi ini memodifikasi variabel nilai yang ada di luar fungsi. Selain itu, jika kita panggil fungsi 'tambah_nilai' beberapa kali, isi dari variabel nilai akan mengalami perubahan incremental (tidak tetap). Sebagai pembuktian, coba jalankan code yang sama berikut sekali lagi! Hasilnya akan berbeda dengan output code sebelumnya (perhatikan pada bagian nilai matkul Mobile). Itulah mengapa fungsi ini bukan termasuk pure function.

<p align="justify"> Bagaimana cara agar fungsi ‘tambah_nilai’ menjadi pure function? Untuk membuat pure function, jangan lupa untuk menghindari penggunaan variabel global atau penggunaan data di luar fungsi. Sebagai gantinya:

1. Fungsi harus menerima ‘nilai’ sebagai parameter
2. Fungsi harus mengembalikan objek ‘nilai’ yang telah dimodifikasi tanpa merubah data luar
3. Caranya adalah dengan membuat salinan (copy) objek ‘nilai’ untuk dimodifikasi kemudian dikembalikan menjadi output fungsi

# **3. Declarative Function**

<p align="justify"> Dalam paradigma pemrograman deklaratif, kita akan ditekankan pada kalimat:
<b>“Pemrograman deklaratif berfokus pada apa (what) dibanding bagaimana (how)”.</b>
Maksudnya adalah, dalam pemrograman deklaratif, kita mendeklarasikan tujuan atau apa (what) hasil yang ingin kita capai, dibandingkan memikirkan bagaimana (how) langkah spesifik untuk mencapainya. Pemrograman fungsional adalah salah satu pendekatan pemrograman deklaratif yang populer.


Untuk memahami pemrograman deklaratif lebih lanjut, mari kita bandingkan dengan lawan dari paradigma pemrograman ini, yaitu pemrograman imperatif.


- <p align="justify"> <b>Pemrograman Imperatif</b>: Menulis program komputer secara detail <b>bagaimana</b> melakukan sesuatu dengan langkah-langkah detail proses yang jelas. Perhatikan potongan kode berikut:




```python
numbers = [1, 2, 3, 4, 5]
double_numbers = []
for number in numbers:
    double_numbers.append(number * 2)
print(double_numbers)
```

<p align="justify"> Dalam program di atas, kita melihat langkah-langkah detail tentang bagaimana mencapai hasil, mulai dari menyiapkan variabel kosong <code>`double_numbers`</code> untuk menyimpan hasil, mendefinisikan loop yang mengiterasi list <code>`numbers`</code>, dan menambahkan hasil dari setiap elemen nya dikali dua ke dalam list <code>`double_numbers`</code>. Pendekatan ini lebih mendetail tentang <b>bagaimana hasil dicapai,</b> yang merupakan karakteristik dari pemrograman imperatif.

- <p align="justify"> Pemrograman Deklaratif: Mendeskripsikan suatu hal (program) tanpa peduli bagaimana cara mendapatkannya. Cukup fokus pada <b>apa</b> yang harus dicapai tanpa perlu tahu bagaimana cara mencapainya. Contoh:

```python
# Mempersiapkan fungsi deklaratif

'''
desc: fungsi double(x) untuk merubah input x menjadi 2x
pre-cond: input x merupakan sembarang data, bisa int, float, dsb
post-cond: 2x sebagai return value
'''
def double(x):
  # Banyak cara untuk mendapatkan 2x. Terserah kita mau pakai rumus apa.
  # Hal ini tidak perlu dijelaskan dalam deskripsi fungsi.
  return x + x
```

<p align="justify">Sebelum menjalankan kode program berikut, pastikan kalian telah menjalankan (run) program imperatif sebelumnya. Jika tidak, akan terjadi <i>error: name 'numbers' is not defined</i>. Karena data <code>`numbers`</code> berada disana.



```python
# Memanfaatkan fungsi deklaratif untuk membangun program

double_numbers = map(double, numbers)
print(list(double_numbers))
```

<p align="justify"> Logika yang sama, bahasa yang sama, dan output yang sama pula,
namun kali ini lebih ringkas. Dalam program di atas, kita cukup fokus pada
apa yang kita inginkan, yaitu memetakan (mapping) setiap elemen dari list
numbers menjadi 2 kalinya dengan menggunakan fungsi <i>map</i> dan <i>double</i>.
Disini kita <b>tidak mendetailkan langkah-langkah</b> iterasi <b>secara eksplisit</b>
untuk mencapai hasil ini. Kita cukup tau <i>apa</i> yang dilakukan oleh fungsi
<i>map</i> dan <i>double</i>., tanpa perlu tau <i>bagaimana</i> cara mereka melakukannya.
Lebih detail tentang fungsi <i>map</i> akan kita bahas pada modul selanjutnya.

<p align="justify"> Demikian juga dengan fungsi <i>double</i> yang kita deklarasikan di awal.
Setiap fungsi yang dibangun dalam paradigma fungsional harus dilengkapi dengan deskripsi fungsi, input apa yang dibutuhkan, dan bentuk output
yang dihasilkan. Hal ini membuat kode lebih mudah dibaca dan dipahami,
serta lebih deklaratif.

# **4. Iterator dan Generator**

## **4.1 Iterator**

<p align="justify">Suatu proses yang diulang lebih dari sekali dengan menerapkan logikayang sama disebut dengan <b>Iterasi</b>. Dalam bahasa pemrograman Python, sebuah loop dibuat dengan beberapa kondisi untuk melakukan iterasi hingga memenuhi batas. Sedangkan <b>Iterator</b> itu sendiri merupakan sebuah objek yang berisi sejumlah nilai yang dapat dihitung dan digunakan untuk mengiterasi objek yang dapat diubah seperti list, tuple, set, dan lain-lain.

<p align="justify"> Iterator diimplementasikan menggunakan class dan variabel lokal untuk iterasi. Iterator mengikuti konsep <i><b>lazy evaluation</b></i>, dimana evaluasi dari ekspresi akan ditahan dan disimpan dalam memori hingga item tersebut dipanggil secara spesifik, yang membantu kita untuk menghindari evaluasi berulang.

Dalam menggunakan iterator:

- <p align="justify">Fungsi __init__() sebagai constructor, tempat kita menginisialisasikan nilai
awal seperti limit. Jika tidak dibutuhkan inisialisasi nilai awal dalam iterator,
maka fungsi ini boleh ditiadakan.
- <p align="justify">Fungsi __iter__() yang digunakan untuk membuat sebuah iterator dengan
sebuah iterable object (self). Fungsi ini juga digunakan untuk melakukan
operasi (menginisiasi dan lain-lain), tetapi harus selalu kembali ke objek
iterator itu sendiri (return self).
- <p align="justify"> Fungsi __next__() digunakan untuk memanggil elemen selanjutnya dalam
iterable object.
- <p align="justify"> Setelah objek iterable selesai diiterasi menggunakan iterator, untuk menggunakannya kembali, objek tersebut perlu ditugaskan kembali (reassign) ke iterator baru. Ini karena <b>iterator hanya bisa digunakan sekali</b>; setelah selesai, ia tidak bisa di-reset atau digunakan ulang secara langsung.


```python
class EvenNumbers:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= self.limit:
            result = self.current
            self.current += 2
            return result
        else:
            raise StopIteration

# Membuat objek iterator
even_iter = EvenNumbers(10)

# Menggunakan iterator untuk mengiterasi dan mencetak bilangan genap
for num in even_iter:
    print(num)
```

<p align="justify"> Sebagian besar objek di Python seperti list, tuple, string, dan lain-lain
adalah iterable. Namun tidak semua yang iterable merupakan iterator, meski bisa
diubah menjadi iterator. Bingung ya? lebih detailnya akan kita bahas selanjutnya
dalam topik lazy evaluation.

<p align="justify"> Suatu objek dikatakan iterable jika dari objek tersebut bisa kita buat
iterator. Caranya sama seperti melakukan casting, yakni dengan memanggil fungsi
iter(). Contohnya sebagai berikut:

```python
# Mendefinisikan list
list_genap = [2, 4, 6, 8, 10]
print(list_genap)
print(type(list_genap))

# Membuat iterator dengan iter()
iterator_genap = iter(list_genap)
print(iterator_genap)
print(type(iterator_genap))
```

<p align="justify"> Iterator menggunakan konsep OOP (Object Oriented Programming) karena
sesuai dengan definisi yang sudah disebutkan di awal: <b>iterator merupakan
sebuah objek yang berisi sejumlah nilai yang dapat dihitung dan digunakan
untuk mengiterasi objek yang dapat diubah</b>. Hal ini tentunya sangat
bertentangan dengan konsep Pemrograman Fungsional yang berfokus pada
penggunaan fungsi (function) sebagai komponen utama dalam menulis program.
Oleh karena itu, kita menggunakan pendekatan lain dalam Python yang lebih
berorientasi pada fungsi dan lebih sesuai dengan paradigma Pemrograman
Fungsional, yaitu <b>Generator</b>.

## **4.2 Generator**

<p align="justify"> Generator adalah jenis khusus dari fungsi pada Python yang menghasilkan
nilai secara malas atau <i><b>lazy evaluation</i></b> dan secara iteratif, daripada
mengembalikan satu nilai secara langsung seperti fungsi biasa.

<p align="justify"> Generator memakai kata kunci <b>‘yield’</b> untuk mengembalikan nilai, yang
membuatnya sangat berguna untuk menghasilkan urutan nilai dalam suatu
konteks dimana perlu menghemat memori atau menangani data secara berurutan.
Untuk pemanggilan objek generator dapat dilakukan dengan method <b>next()</b>
seperti pada iterator. Perhatikan contoh dibawah ini:


<p align="justify"> Pertama-tama kita perlu membuat sebuah fungsi generator dengan
menambahkan keyword <b>‘yield’</b> sebagai pengganti keyword return. Jangan lupa
untuk menambahkan deklarasi/deskripsi pada setiap fungsi yang kita bangun!


```python
"""
desc: fungsi generator 'bilangan_genap(limit)' akan menghasilkan bilangan genap
      dari 0 hingga 'limit'.
pre-cond: 'limit' harus berupa int >= 0
post-cond: fungsi 'bilangan_genap' menghasilkan object generator yang berisikan
           bilangan genap dari 0 sampai 'limit'
"""

def bilangan_genap(limit):
    n = 0
    while n <= limit:
        yield n
        n += 2
```

Pada kode di atas, fungsi ‘bilangan_genap(limit)’ adalah fungsi generator
yang menghasilkan object generator yang berisikan bilangan genap dari 0 sampai
'limit'. Keyword ‘yield’ digunakan untuk mengembalikan nilai secara bertahap
setiap kali data dari object generator dipanggil tanpa menghentikan eksekusi
fungsi sepenuhnya.



Kemudian kita siapkan sebuah variabel `generator_genap` sebagai keranjang penyimpanan hasil dari fungsi `bilangan_genap`:



! Pastikan kalian sudah *run* kode fungsi `bilangan_genap` sebelum *run* kode berikut! Agar tidak error.









```python
generator_genap = bilangan_genap(10)

print(type(bilangan_genap))
print(type(generator_genap))
print(bilangan_genap)
print(generator_genap)
```

<p align="justify"> <code>`generator_genap`</code> adalah objek generator yang dihasilkan dengan memanggil fungsi <code>`bilangan_genap(10)`</code>. Kalian bisa membedakan keduanya dengan cara memanggil fungsi <code>`print`</code> dan <code>`type`</code>



<p align="justify"> Object generator tidak dapat dicetak secara langsung karena dia <b><i>lazy</b></i>. Kita perlu melakukan iterasi/loop untuk mendapatkan isinya satu persatu atau menggunakan method next() seperti berikut:

```python
print (next(generator_genap))
print (next(generator_genap))

for num in generator_genap: #looping akan melanjutkan proses generate dari angka genap terakhir
    print(num,end = ' ')

print (next(generator_genap))  #eror disini karena objek generator telah habis/mencapai limit
```

Disini kita menggunakan fungsi <code>`next(generator_genap)`</code> untuk mengambil nilai berikutnya dari generator secara berurutan. Loop <code>`for num in generator_genap`</code> juga dapat digunakan untuk mencetak setiap bilangan genap selanjutnya yang dihasilkan oleh generator.

<p align="justify">Setelah program dieksekusi, maka akan dihasilkan angka genap kurang dari limit (10), yaitu: 0, 2, 4, 6, dan 8. Perintah <code>`next()`</code> berikutnya akan memunculkan exception <code>`StopIteration`</code>, yang mana artinya elemen berikutnya dari generator tersebut telah habis dan tidak ada lagi nilai yang bisa dihasilkan. Objek generator pun tidak bisa digunakan lagi.

## **4.3 Generator Expression**

<p align="justify"> Generator dapat dibuat dengan cara yang lebih simple tanpa harus
membuat fungsi yang menggunakan kata kunci 'yield' terlebih dahulu. Hal ini
dikenal dengan sebutan <b>Generator Expression</b>. Cara membuatnya cukup dengan
menggunakan tanda kurung dan perulangan didalamnya (bisa juga ditambahkan
kondisi if). Perhatikan contoh berikut:

```python
generator_genap = (i for i in range (21) if i % 2 == 0)
print(generator_genap)
```

<p align="justify"> Kode di atas adalah cara yang sama untuk membuat iterasi dengan range 21 yang akan mencetak angka genap dari 0 sampai 20. Lebih sederhana lagi daripada harus membuat fungsi generator yang panjang seperti sebelumnya ataupun class iterator yang rumit bukan...Inilah kekuatan dari fungsional programming yang deklaratif.

# **5. Lazy Evaluation**

<p align="justify"> Dari tadi istilah ini sudah berulang kali disebutkan. Tapi apa sih sebenarnya maksudnya? Daripada banyak teori bertele-tele, sebaiknya kita langsung praktek saja biar lebih paham tentang apa itu lazy evaluation. Pertama-tama, mari kita lihat kode di bawah:

```python
ini_range = range(1, 100)
ini_list = list(range(1, 100))
```

<p align="justify"> Kode di atas merupakan contoh penggunaan range dan list dalam python.Sekilas memang keduanya mirip, tetapi ternyata terdapat perbedaan antarakeduanya. Pada range, kita menggunakan konsep <b>lazy evaluation</b>, di mana evaluasiekspresi akan menghasilkan satu item pada satu waktu dan menghasilkan item hanyapada saat dibutuhkan. Item dari objek yang lazy tidak bisa diakses secara langsung.Seringkali kita butuh iterasi/loop untuk mengaksesnya satu persatu. Silahkan kaliancoba print keduanya untuk membuktikan!




```python
print("data range = ",ini_range)
print("data list = ",ini_list)
```

<p align="justify"> Dengan demikian kita bisa menganggap bahwa penggunaan range yang lazy pasti akan lebih hemat memori daripada list, mengingat data yang tidak bisa diakses berarti data tersebut belum tersedia di memory internal kita. Mari kita buktikan dengan menjalankan potongan kode berikut:

```python
from sys import getsizeof
x = getsizeof(ini_range)
print("Size dari range: ", x)

y = getsizeof(ini_list)
print("Size dari list:", y)
```

Namun, apakah kemudian objek range yang lazy bisa lebih hemat waktu eksekusinya juga dibanding dengan list? Mari kita buktikan:

```python
import timeit
print(timeit.timeit('''r = range(1, 100)''', number = 1000000))
print(timeit.timeit('''lst = list(range(1, 100))''', number = 1000000))
```

<p align="justify"> Ternyata, waktu eksekusi range lebih singkat daripada list. Dapat disimpulkan
bahwa range <b>lebih hemat memori serta hemat waktu daripada list</b>. Hal ini
membuktikan bahwa sesuatu yang malas tidak selalu berkonotasi negatif. Khususnya
pada pemrograman fungsional. Yang disebut lazy disini malah bisa bekerja lebih
cepat dan lebih hemat. Konsep lazy evaluation merupakan salah satu teknik unggulan
dalam paradigma fungsional. Keren kan fungsional...
